# LINE & SCATTER Plot

### Line Plot

```python
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
y = [71940, 19000, 7500, 5000, 8900, 12100, 25000, 11163, 12778, 98042, 11900, 14253]
```

### Comparision

```python
x = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
y = [71940, 19000, 7500, 5000, 8900, 12100, 25000, 11163, 12778, 98042, 11900, 14253]

x2 = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
y2 = [12000, 15000, 14323, 13500, 16400, 94000, 17800, 12311, 20319, 21341, 21000, 25000 ]
```

```python
plt.plot(x, y, color='b', linewidth= 4)
plt.plot(x2, y2, color='r', linewidth= 4)

plt.title('Comparing Hotti Family Expenses of Two Years')
plt.ylabel('Expense')
plt.xlabel('Months')

plt.annotate(xy=['Jan', 71940], text = '2020')
plt.annotate(xy=['Jan', 9000], text = '2021')

plt.legend(['2020', '2021'], loc=1)

plt.show
```
