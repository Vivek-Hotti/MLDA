Formula Used for Normalizing Values using the MinMaxScaler

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
# initializing the function
```

Now if we need to perform pre-processing of data. 

### <b>Data Processing</b>
- dropping categorical data.  <i>(in this case: column Ocean_Proximity)</i>
- dropping rows with Null Values 

```python
data = data.drop(data.columns[[9]], axis = 1)
# dropping entire 10 th column (.i.e 9th index)
```

```python
data = data.dropna(axis = 0)
# dropping entire rows where Null value is present
```

```python
data.head(10)
```

### Using the Scaling (that was initalized earlier)

```python
scaler.fit(data)
# passing above pre-processed data into scaler function which was initalized as the MinMaxScaling function earlier at cell 5.
```

```python
scaled_data = pd.DataFrame(data = scaler.transform(data), columns = data.columns, index = data.index)
# new_dataframe_name=pd.DataFrame(data=sclaer.transform(datasetName), columns=data.columns, index-data.index)
```
