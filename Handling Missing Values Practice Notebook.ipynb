## <font color='green'>Approach 1: Using the fillna() method </font>

#### <b>The fillna() function is used only when all the NULL values in every column has to be filled with the same data / value.</b>

As you will see in this example, we have filled every Null value with the number <b>999</b>

```python
# seprately selecting and holding data that has got NULL Values.
cols = ['Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'YearBuilt', 'CouncilArea', 'Lattitude', 'Longtitude', 'Regionname', 'Propertycount']
```

```python
data[cols].isnull().sum()
```

```python
data[cols].head()
```

```python
data[cols].describe()
```

```python
a = data[cols].fillna(value = 999)
```

```python
a.head(5)
```

```python
a.isnull().sum()
```

## <font color='green'>Approach 2: Using Imputer </font>

Strategy to handle missing values :
- mean
- median
- most_frequent (mode)

```python
from sklearn.impute import SimpleImputer

imputer_mean = SimpleImputer(missing_values = np.nan, strategy = "mean")
imputer_median = SimpleImputer(missing_values = np.nan, strategy = "median")
```
